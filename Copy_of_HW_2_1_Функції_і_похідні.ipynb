1. Задана функція залежності оцінки від кількості годин, витрачених на навчання: $ f(x) = 10x - 5 $.
  1. З допомогою Python обчисліть похідну цієї функції.
  2. Використовуйте бібліотеку `matplotlib` для побудови графіків функції та її похідної в діапазоні значень від 0 до 10 годин. Проаналізуйте, як зміна кількості годин вивчення впливає на оцінку.



```python
import sympy as sp

# Визначаємо символічну змінну
x = sp.symbols('x')

# Лінійна функція
f_lin = 10 * x - 5
derivative_f_lin = sp.diff(f_lin, x)

print("Похідна лінійної функції 10 * x - 5:", derivative_f_lin)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Визначення функції та її похідної
x = np.linspace(0, 10, 100)
f = 10 * x - 5
df = np.full(x.shape, 10) #  Creating an array with the same shape as x and filled with the value 10


# Створення графіку
plt.figure(figsize=(10, 5))
# Графік функції
plt.plot(x, f, label='f(x) = 10 * x - 5', color='blue')
# Графік похідної
plt.plot(x, df, label="f'(x) = 10", color='green', linestyle='--')

# Додавання легенди, міток та заголовка
plt.title("Графік функції та її похідної")
plt.xlabel("Кількість годин  x")
plt.ylabel("Оцінка y")
plt.legend()
plt.grid(True)

plt.show()
```

Графік функції показує, що чим більше годин вивчення, тим вища оцінка. Оцінка збільшується на 10 за кожну годину навчання.
Похідна показує, що швидкість зміни оцінки є сталою: зростає на 10 одиниць за кожну додаткову годину навчання.

2. Використовуючи Python і `SymPy`, обчисліть похідні для функцій $ f(x) = \sin(x) $ та $ f(x) = \ln(x) $. Зобразіть ці функції та їхні похідні за допомогою `matplotlib`.



```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt

# Визначаємо символічну змінну
x = sp.symbols('x')

# Функції
sin_x = sp.sin(x)
ln_x = sp.ln(x)

# Похідні
derivative_sin_x = sp.diff(sin_x, x)
derivative_ln_x = sp.diff(ln_x, x)

print("Похідна функції f(x) = sin(x):", derivative_sin_x)
print("Похідна функції f(x) = ln(x):", derivative_ln_x)

# Конвертація символічної функції у числову для роботи з numpy
fsin = sp.lambdify(x, sin_x, 'numpy')   # f(x) = sin(x)
dfsin = sp.lambdify(x, derivative_sin_x, 'numpy') # f'(x) = cos(x)
fln = sp.lambdify(x, ln_x, 'numpy')   # f(x) = ln(x)
dfln = sp.lambdify(x, derivative_ln_x, 'numpy') # f'(x) = 1/x

# Числові значення x
x_vals_sin = np.linspace(-2 * np.pi, 2 * np.pi, 500)  # Для sin(x)
x_vals_ln = np.linspace(0.1, 10, 500)  # Для ln(x), x > 0

# Створення графіку
plt.figure(figsize=(12, 6))

# Графік функції та похідної sin(x)
plt.subplot(1, 2, 1)
plt.plot(x_vals_sin, fsin(x_vals_sin), label='f(x) = sin(x)', color='blue')
plt.plot(x_vals_sin, dfsin(x_vals_sin), label=f"f'(x) = cos(x)", color='orange', linestyle='--')
plt.title("Графік sin(x) та його похідної")
plt.xlabel("x")
plt.ylabel("y")
plt.axhline(0, color='black', linestyle='--', linewidth=0.8)
plt.axvline(0, color='black', linestyle='--', linewidth=0.8)
plt.legend()
plt.grid()

# Графік функції та похідної ln(x)
plt.subplot(1, 2, 2)
plt.plot(x_vals_ln, fln(x_vals_ln), label='f(x) = ln(x)', color='green')
plt.plot(x_vals_ln, dfln(x_vals_ln), label=f"f'(x) = 1/x", color='red', linestyle='--')
plt.title("Графік ln(x) та його похідної")
plt.xlabel("x")
plt.ylabel("y")
plt.axhline(0, color='black', linestyle='--', linewidth=0.8)
plt.axvline(0, color='black', linestyle='--', linewidth=0.8)
plt.legend()
plt.grid()

plt.tight_layout()
plt.show()
```


3. Загальні витрати компанії від обсягу виробництва задані функцією $ f(x) = x^2 - 5x + 1 $, де $ x $ — кількість вироблених одиниць. Обчисліть похідну цієї функції з допомогою SymPy та візуалізуйте значення функції і її похідної на відрізку від 0 до 100.

```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt

# Визначаємо символічну змінну та функцію
x = sp.symbols('x')
f = x**2 - 5*x + 1

# Обчислюємо похідну
derivative_f = sp.diff(f, x)

# Виводимо формулу похідної
print("Похідна функції витрат f(x) = x**2 - 5*x + 1:   ", derivative_f)

# Конвертація символічної функції у числову для роботи з numpy
f_numeric = sp.lambdify(x, f, 'numpy')        # f(x) = x^2 - 5x + 1
df_numeric = sp.lambdify(x, derivative_f, 'numpy')  # f'(x)

# Числовий діапазон для x
x_vals = np.linspace(0, 100, 500)  # Від 0 до 100

# Створення графіка
plt.figure(figsize=(8, 5))
plt.plot(x_vals, f_numeric(x_vals), label='f(x) = x^2 - 5x + 1', color='blue')
plt.plot(x_vals, df_numeric(x_vals), label="f'(x) = 2x - 5", color='orange', linestyle='--')
plt.title("Графік f(x) та його похідної")
plt.xlabel("Кількість вироблених одиниць (x)")
plt.ylabel("Витрати y")
plt.axhline(0, color='black', linestyle='--', linewidth=0.8)
plt.axvline(0, color='black', linestyle='--', linewidth=0.8)
plt.legend()
plt.grid()
plt.show()
```

Витрати значно збільшуються з ростом обсягу виробництва.
